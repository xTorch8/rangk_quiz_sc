## Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([2, 4, 6, 8, 10, 12, 14])
```

#### Least Square Regression

```python
A = np.vstack([x, np.ones(len(x))]).T
A
```

```python
y = y[:, np.newaxis]
y
```

```python
pinv = np.linalg.pinv(A)
coeff = pinv.dot(y)
coeff
```

```python
print(f"y = {round(coeff[0][0], 3)}x + {round(coeff[1][0], 3)}")
```

```python
plt.plot(x, x * coeff[0] + coeff[1])
plt.plot(x, y, ".")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Relationship Between x and y")

plt.show()
```
